Pokemon Import

```python
import pandas as pd

import numpy as np

pkm_df = pd.read_csv('https://raw.githubusercontent.com/sziccardi/MLCamp2024_DataRepository/main/pokemon.csv')
```

Cleaning Data

```python
pkm2_df = pkm_df.drop(columns=['abilities','classfication', 'japanese_name','name','pokedex_number','height_m', 'type1','type2','generation','weight_kg'], axis=1)

pkm2_df.columns

pkm2_df['capture_rate'].replace('30 (Meteorite)255 (Core)', 30, inplace=True)

pkm2_df['capture_rate'] = pkm2_df['capture_rate'].astype(str).astype(int)

pkm3_df = pkm2_df
```

```python
pkm3_df.dtypes

print(pkm3_df.columns)
```

```python
pkm3_df = pkm3_df.fillna(value=-1)
pkm3_df = pd.get_dummies(pkm3_df, columns=['percentage_male'], dummy_na=True)
print(pkm3_df.columns)

pkm3_df.describe(include='all')
```

Plot

```python
print(pkm3_df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.pairplot(pkm_df, vars=['attack','base_egg_steps', 'base_happiness', 'base_total'], hue = 'is_legendary')

plt.show()
```

```python
plt.figure()
sns.pairplot(pkm3_df, vars=['experience_growth', 'hp','sp_attack', 'sp_defense', 'speed','capture_rate'], hue = "is_legendary")
plt.show()
```

Test-Training Data

```python
from sklearn.model_selection import train_test_split

features = ['attack', 'base_egg_steps',
       'base_happiness', 'base_total', 'capture_rate', 'defense',
       'experience_growth', 'hp', 'sp_attack', 'sp_defense', 'speed',
       'percentage_male_-1.0', 'percentage_male_0.0',
       'percentage_male_11.2', 'percentage_male_24.6', 'percentage_male_50.0',
       'percentage_male_75.4', 'percentage_male_88.1', 'percentage_male_100.0',
       'percentage_male_nan']

X = pkm3_df[features]
y = pkm3_df['is_legendary']

X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)

pkm3_df.describe(include='all')
```

K Nearest Neighbor

```python
#Train your nearest neighbor classifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

classifier = KNeighborsClassifier(n_neighbors=17)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

y_train_pred = classifier.predict(X_train)
print(accuracy_score(y_train, y_train_pred))


#loop or smtg

k_list = list(range(1,50,2))
cv_scores = []

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k,weights='distance') #can change between distance and uniform points
    scores = cross_val_score(knn, X_train, y_train, cv = StratifiedKFold(n_splits=10, shuffle=True), scoring = 'accuracy')
    cv_scores.append(scores.mean())

plt.figure()
plt.title('Performance of Uniform K Nearest Neighbors Algorithm')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Accuracy Score')
plt.plot(k_list, cv_scores)

plt.show()

best_k = k_list[cv_scores.index(max(cv_scores))]
print(best_k)
```

New K-nearest neighbor / getting rid of "against_type"

```python
from sklearn.model_selection import train_test_split

features3 = ['attack', 'base_egg_steps',
       'base_happiness', 'base_total', 'capture_rate', 'defense',
       'experience_growth', 'hp', 'sp_attack', 'sp_defense', 'speed',
       'percentage_male_-1.0', 'percentage_male_0.0',
       'percentage_male_11.2', 'percentage_male_24.6', 'percentage_male_50.0',
       'percentage_male_75.4', 'percentage_male_88.1', 'percentage_male_100.0',
       'percentage_male_nan']

X = pkm3_df[features3]
y = pkm3_df['is_legendary']

X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)

X.describe(include='all')

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

classifier = KNeighborsClassifier(n_neighbors=23)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

y_train_pred = classifier.predict(X_train)
print(accuracy_score(y_train, y_train_pred))


#loop or smtg

k_list = list(range(1,50,2))
cv_scores = []

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k,weights='distance') #can change between distance and uniform points
    scores = cross_val_score(knn, X_train, y_train, cv = StratifiedKFold(n_splits=10, shuffle=True), scoring = 'accuracy')
    cv_scores.append(scores.mean())

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
plt.title('Performance of Uniform K Nearest Neighbors Algorithm')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Accuracy Score')
plt.plot(k_list, cv_scores)

plt.show()

best_k = k_list[cv_scores.index(max(cv_scores))]
print(best_k)
```
